```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Values':[0,70,30,90,60], # Added a fifth value to match the length of other columns
    'Sales':np.random.randint(100,500,5),
    'Profit':np.random.randint(10,100,5)
})

display(df)
```

```python
import matplotlib.pyplot as plt

#1. Area plot

plt.figure()
plt.title('Rudransh ')
plt.fill_between(df['Category'],df['Values'],color='green',alpha=1)
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
#2. Two area plots in single graph

import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")   # apply seaborn style
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'], df['Sales'],
                 color='skyblue', alpha=0.5, label='Sales')

plt.fill_between(df['Category'], df['Profit'],
                 color='orange', alpha=0.5, label='Profit')

plt.title("rudransh")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2. Pie Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%')#autopct is a command for decimals
plt.title("Rudransh")
plt.show()
```

```python
# 3. Donut Chart
plt.figure() #it opens in a new window
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Rudransh")
plt.show()

#gcf=Get current figure gca=Get Current Axes
```

```python
#4. Boxplot (Outliers Detection)#plotted between one varaible box plot is used to find outliers
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Rudransh")
plt.show()
```

```python
#5. Heatmap (Correlation) #yey correlation finout karne kai kaam aata hai

plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm') #if annot is false than no value would be shown
plt.title("Rudransh")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Values','Sales','Profit']].corr()

plt.figure(figsize=(5,4))

sns.heatmap(corr, annot=True)

plt.title("Rudransh")

plt.show()
```

```python
#Bubble Plot (3 Variable Chart)
import matplotlib.pyplot as plt

plt.scatter(df['Sales'], df['Profit'],
            s=df['Values']*10)   # size of bubbles

plt.title("rudransh")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(50, 300),      # min-max size
                palette='viridis')

plt.title("rudransh")

plt.show()
```

```python
#2. Second Dataset
np.random.seed(7)
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())
```

```python
#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("rudransh (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#3. Histogram (Credit Score Distribution)
plt.hist(df['Credit_Score'], bins=6, color='skyblue', edgecolor='black')

plt.title("rudransh")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.show()
```

```python
#Bubble Plot (Advanced Scatter)
plt.scatter(df['Income'], df['Loan_Amount'],
            s=df['Credit_Score']/2,
            c=df['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("rudransh (Income vs Loan)")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("rudransh")

plt.show()
```
